## Removing html < > Tags

```python
import re
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean,'',text)
```

```python
data['review']=data['review'].apply(clean_html)
```

```python
data
```

```python
data['review'].iloc[3]
```

## Removing Special Character

```python
def remove_special(text):
    x = ''
    for i in text:
        if i.isalnum():
            x = x+i
        else:
            x = x+' '
    return x        
```

```python
data['review']=data['review'].apply(remove_special)
```

```python
data['review'].iloc[3]
```

## Convert To Lowercase

```python
def convert_lower(text):
    return text.lower()
```

```python
data['review']=data['review'].apply(convert_lower)
```

```python
data['review'].iloc[3]
```

## Removing Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
a = stopwords.words('english')
a
```

```python
def remove_stopwords(text):
    x = []
    for i in text.split():
        if i not in a:
            x.append(i)
        else:
            pass
    return x        
```

```python
data['review']=data['review'].apply(remove_stopwords)
```

```python
data['review'].iloc[3]
```

## Stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
    x = []
    for i in text:
        x.append(ps.stem(i))
    return x    
```

```python
data['review']=data['review'].apply(stem_words)
```

```python
data['review'].iloc[3]
```

## Joining

```python
def join(list_input):
    return ' '.join(list_input)
```

```python
data['review']=data['review'].apply(join)
```

```python
data['review'].iloc[3]
```

# APPLYING MACHINE LEARNING ALGOS

```python
X = data['review']
y = data['sentiment']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
cv.fit_transform(data['review'])
```

```python
X = cv.fit_transform(data['review']).toarray()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
nb = GaussianNB()
```
